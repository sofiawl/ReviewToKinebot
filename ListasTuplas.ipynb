Um compreensão de lista consiste de um par de colchetes contendo uma expressão seguida de uma cláusula for, e então zero ou mais cláusulas for ou if. O resultado será uma nova lista resultante da avaliação da expressão no contexto das cláusulas for e if.

# Lista mutável

O perigo de uma lista ou objeto mutável. 
O tamanho e valores de uma lista mudam porque não se sabe se é uma nova lista ou antiga lista, por isso precisamo esclarecer quando é uma nova lista.

```python
def listaMutavel(lista = list()):
    
    print(len(lista))
    print(lista)
    lista.append(3)    
```

```python
def listaNaoMutavel(lista = None):
    if lista == None:
        lista = list()

    print(len(lista))
    print(lista)
    lista.append(3)
```

```python
print('Lista mutável:')
listaMutavel()
listaMutavel()
print('Lista NÃO mutável:')
listaNaoMutavel()
listaNaoMutavel()
```

# Lista Mutável de Objetos


```python
class contaCorrente:
    def __init__ (self, codigo):
        self.codigo = codigo
        self.saldo = 0

    def deposito(self, valor):
        self.saldo += valor

    def __str__ (self):
        return "[>>Código {}  Saldo {}<<]".format(self.codigo, self.saldo)
```
