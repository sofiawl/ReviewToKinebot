# Precificação Inteligente

Trata-se de um algoritmo capaz de analisar as características de um imóvel — como comodidade, tamanho, ocupação do espaço em determinado período — e sugerir à pessoa anfitriã um preço a ser cobrado por diárias que garanta o lucro em momentos de alta demanda.

```python
import pandas as pd

data = pd.read_json('dados_hospedagem.json')
data.head()
```

```python
# Só é possível normalizar dados bem aninhados
data = pd.json_normalize(data['info_moveis'])
data
```

```python
columns = list(data.columns)
columns
```

```python
# Desagrupando somente dados em Listas

data = data.explode(columns[3:])
data
```

```python
data.reset_index(inplace=True, drop=True)
data.head()
```

```python
data.info()
```

```python
import numpy as np

data['max_hospedes'].astype(np.int8)
```

```python
data['max_hospedes'] = data['max_hospedes'].astype(np.int8)
```

```python
data.info()
```

```python
integers_col = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']

data[integers_col] = data[integers_col].astype(np.int8)

data.info()
```

```python
# Erro por causa dos caracteres . e $

#reals_col = ['avaliacao_geral', 'preco']

#data[reals_col] = data[reals_col].astype(np.float32)
```

```python
data['preco']
```

```python
data['preco'] = data['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
data['preco']
```

```python
data['preco'] = data['preco'].astype(np.float32)
```

```python
data['avaliacao_geral']
```

```python
data[['taxa_deposito', 'taxa_limpeza']]
```

```python
data[['taxa_deposito', 'taxa_limpeza']] = data[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())

data[['taxa_deposito', 'taxa_limpeza']]
```

```python
data[['taxa_deposito', 'taxa_limpeza']] = data[['taxa_deposito', 'taxa_limpeza']].astype(np.float32)

data[['taxa_deposito', 'taxa_limpeza']]
```

```python
data.info()
```

```python
# OBS: Tratar valor e tirar nan

data['avaliacao_geral'] = data['avaliacao_geral'].apply(lambda x: x.replace('nan', '').strip())

data['avaliacao_geral']
```

```python
data['descricao_local']
```

```python
data['descricao_local'] = data['descricao_local'].str.lower()

data['descricao_local']
```

```python
data['descricao_local'] = data['descricao_local'].replace('[^a-zA-Z0-9\-\']', ' ', regex=True)

data['descricao_local']
```

```python
# (?<!\w) lookbehind negativa: se não houver caractere antes do hífen é "True"
# - : e lógico
# (?!\w)  lookhead negativa: se há um caractere de palavra após hífen é "True"



data['descricao_local'].replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
data['descricao_local'] = data['descricao_local'].str.split()

data['descricao_local']
```

```python
data.head()
```

```python
data['comodidades'] = data['comodidades'].replace('[\{|}|\"]', '', regex=True)

data['comodidades']
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.lower()
data['descricao_vizinhanca']
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
data['descricao_vizinhanca']
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].replace('(?<!\w)-(?!\w)', ' ', regex=True)

data['descricao_vizinhanca']
```

```python
data['descricao_local'] = data['descricao_vizinhanca'].str.split()

data['descricao_local']
```
