### Lista mutável

O perigo de uma lista ou objeto mutável. 
O tamanho e valores de uma lista mudam porque não se sabe se é uma nova lista ou antiga lista, por isso precisamo esclarecer quando é uma nova lista.

```python
def listaMutavel(lista = list()):
    
    print(len(lista))
    print(lista)
    lista.append(3)    
```

```python
def listaNaoMutavel(lista = None):
    if lista == None:
        lista = list()

    print(len(lista))
    print(lista)
    lista.append(3)
```

```python
print('Lista mutável:')
listaMutavel()
listaMutavel()
print('Lista NÃO mutável:')
listaNaoMutavel()
listaNaoMutavel()
```

# Tuplas ()

Tuplas são sequências imutáveis, tipicamente usadas para armazenar coleções de dados heterogêneos (como as tuplas de 2 elementos produzidas pelo função embutida enumerate()). Tuplas também são usadas para casos em que seja necessária uma sequência imutável de dados homogêneos (como permitir o armazenamento em uma instância set ou dict).

```python
contaX = ('Guilherme', 31, 11100)
contaY = ('Dani', 19, 50)
```

```python
contaX.deposita(100)
```

```python
def deposita(conta, valor):
    novoSaldo = conta[2] + valor
    idade = conta[1]
    nome = conta[0]
    return (nome, idade, novoSaldo)
```

```python
print(contaX)
deposita(contaX, 100)
print(contaX)
```

```python
print(contaX)
contaX = deposita(contaX, 100)
print(contaX)
```

```python
usuarios = [contaX, contaY]
usuarios
```

# Polimorfismos e Herança

```python
from abc import ABCMeta, abstractmethod

class Conta(metaclass = ABCMeta):
    def __init__(self, codigo):
        self._codigo = codigo
        self._saldo = 0

    def deposito(self, valor):
        self._saldo += valor

    # Se for criada uma subclasse sem atributo fimMes dá erro
    # É ótimo para deixar o código padrão
    @abstractmethod
    def fimMes(self):
        pass

    def __str__(self):
        return "[>>Código {}  Saldo {}<<]".format(self._codigo, self._saldo)
```

```python
class ContaInvestimento(Conta):
    pass
```

```python
print(Conta(88))
```

```python
ContaInvestimento(71)
```

```python
class ContaCorrente(Conta):
    def fimMes(self):
        self._saldo -= 2

class ContaPoupanca(Conta):
    def fimMes(self):
        self._saldo *= 1.01
        self._saldo -= 3
```

```python
contaW = ContaCorrente(800)
contaZ = ContaPoupanca(76)
contaW.deposito(999)
contaZ.deposito(700)
print(contaW, contaZ)
```

```python
contasGerais = [contaW, contaZ]

for conta in contasGerais:
    conta.fimMes()
    print(conta)
```

# Ordenação sem ordem natural

```python
class contaSalario:
    def __init__(self, codigo):
        self._codigo = codigo
        self._saldo = 0

    def __eq__(self, outro):
        if type(outro) != contaSalario:
            pass

        return self._codigo == outro._codigo and self._saldo == outro._saldo
    
    def deposita(self, valor):
        self._saldo += valor
        
    def __str__(self):
        return '[>> Código {}  Saldo {}<<]'.format(self._codigo, self._saldo)
    
```

```python
contaLuiz = contaSalario(0)
contaLuiz.deposita(7000)
contaAlessandra = contaSalario(1)
contaAlessandra.deposita(100000)
contaVinicius = contaSalario(2)
contaVinicius.deposita(-33)

contas = [contaLuiz, contaAlessandra, contaVinicius]
contas
```

```python
for conta in contas:
    print(conta)
```

```python
# CUIDADO, ESTÁ EXTRAÍNDO VALOR PRIVADO DE UM OBJETO
# UMA MIERDA
def extraiSaldo(conta):
    return conta._saldo

for conta in sorted(contas, key=extraiSaldo):
    print(conta)
```

```python
# O correto:

class contaSalario:
    def __init__(self, codigo):
        self._codigo = codigo
        self._saldo = 0

    def __eq__(self, outro):
        if type(outro) != contaSalario:
            pass

        return self._codigo == outro._codigo and self._saldo == outro._saldo
    
    def deposita(self, valor):
        self._saldo += valor
        
    def __lt__(self, outro):
        return self._saldo < outro._saldo
        
    def __str__(self):
        return '[>> Código {}  Saldo {}<<]'.format(self._codigo, self._saldo)
    
```

```python
contaLuiz = contaSalario(0)
contaLuiz.deposita(7000)
contaAlessandra = contaSalario(1)
contaAlessandra.deposita(100000)
contaVinicius = contaSalario(2)
contaVinicius.deposita(-33)
```

```python
contaVinicius < contaAlessandra
```

```python
for conta in sorted(contas):
    print(conta)
```

# Ordenação completa e functools

```python
from functools import total_ordering

@total_ordering
class contaSalario:
    def __init__(self, codigo):
        self._codigo = codigo
        self._saldo = 0

    def __eq__(self, outro):
        if type(outro) != contaSalario:
            pass

        return self._codigo == outro._codigo and self._saldo == outro._saldo
    
    def __lt__(self, outro):
        if self._saldo != outro._saldo:
            return self._saldo < outro._saldo
        return self._codigo < outro._codigo

    
    def deposita(self, valor):
        self._saldo += valor
        
    def __str__(self):
        return '[>> Código {}  Saldo {}<<]'.format(self._codigo, self._saldo)
    
```

```python
contaH = contaSalario(89)
contaH.deposita(-70)
contaJ = contaSalario(7)
contaJ.deposita(-70)
contaK = contaSalario(999)
contaK.deposita(800)

contas = [contaH, contaJ, contaK]

for conta in sorted(contas):
    print(conta)
```
